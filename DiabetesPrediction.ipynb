## Import dependencies

```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Dataset

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target
```

```python
X = df.drop(['target'], axis=1)
y = df['target']
```

```python
X
```

```python
y
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
X_train.head()
```

```python
sns.heatmap(X_train.corr(),annot=True)
```

#### Import Model


```python
from sklearn.tree import DecisionTreeRegressor

decision_reg = DecisionTreeRegressor()
```

```python
decision_reg.fit(X_train,y_train)
```

#### Hyperparameter Tuning

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV

grids = GridSearchCV(decision_reg,param_grid=param,scoring='accuracy',cv = 5,n_jobs=-1)
```

```python
import warnings
warnings.filterwarnings('ignore')
grids.fit(X_train,y_train)
```

```python
grids.best_params_
```

```python
grids.best_score_
```

```python
y_pred = grids.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

#### Tree visualization

```python
selectedmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='log2',splitter='random')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedmodel,filled=True)
```
